```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
from shapely import wkt
import numpy as np
from shapely.geometry import Point
```

```python
# Load your Toronto shapefile
toronto_map = gpd.read_file('toronto_neighbourhood_shape.shp')
cooling_center_locations = gpd.read_file('toronto_cooling_shape.shp')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
toronto_map.plot(ax=ax, edgecolor='black', facecolor='none')
cooling_center_locations.plot(ax=ax, markersize = 8)


# Overlay your data
# This assumes you've converted your data into a format compatible with matplotlib
# For complex data manipulations, consider rasterio to work with raster data and overlay it
# Plotting directly depends on your dataset's format and may require conversion or rasterization

ax.set_title('Toronto Neighbourhood Map With All Cooling Centers')
plt.savefig('bar.pdf')
plt.show()
```

```python
cooling_center_data = pd.read_csv('cooling_centres_minus_pools.csv')
# cooling_center_data.columns
```

```python
# cooling_center_data['geometry'] = cooling_center_data['geometry'].apply(wkt.loads)
# gdf = gpd.GeoDataFrame(cooling_center_data,  crs ='epsg:4326')
lons = []
lats = []
for i in range(len(cooling_center_data['geometry'])):
    coords = np.array(cooling_center_data['geometry'][i][40:][:-3].split(',')).astype(float)
    lons.append(coords[0])
    lats.append(coords[1])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
toronto_map.plot(ax=ax, edgecolor='black', facecolor='none')
#cooling_center_locations.plot(ax=ax, markersize = 8)


# Overlay your data
# This assumes you've converted your data into a format compatible with matplotlib
# For complex data manipulations, consider rasterio to work with raster data and overlay it
# Plotting directly depends on your dataset's format and may require conversion or rasterization

# help(ax.scatter)
ax.scatter(lons, lats, s = 5)

ax.set_title('Toronto Neighbourhood Map With Cooling Centers (No Pools)')
plt.savefig('foo.pdf')
plt.show()
```
